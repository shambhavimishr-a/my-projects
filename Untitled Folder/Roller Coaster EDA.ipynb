# ASKING QUESTIONS

# Q. What are the locations with fastest roller coaster

```python
df['Location'].value_counts()
```

```python
df.query('Location != "Other"') \
.groupby('Location')['Speed_mph'] \
.agg(['mean','count'])
```

```python
df.query('Location != "Other"') \
        .groupby('Location')['Speed_mph'] \
        .agg(['mean','count']) \
        .query('count >= 10')  \
        .sort_values('mean')
```

```python
ax=df.query('Location != "Other"') \
        .groupby('Location')['Speed_mph'] \
        .agg(['mean','count']) \
        .query('count >= 10')  \
        .sort_values('mean')['mean'] \
        .plot(kind='barh',title ='Average speed by location')

ax.set_xlabel('Average Coaster Speed')
```
